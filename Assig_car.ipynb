```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/PratikPhysics/SNDT-College/refs/heads/main/car_price_dataset.csv")
df
```

```python
#Q.#1. What is the average price of all cars in the dataset?

print(f"The average price of all cars in the dataset is ${round(df['Price'].mean(),2)}")
```

```python
#Q.2 Which car brand has the highest average price?

brand = df[df["Price"] == df["Price"].max()]["Brand"][1100]

print(f"The brand with the highest price is {brand}")
```

```python
#Q.3.What is the most common fuel type in the dataset?
df['Fuel_Type'].mode()
```

```python
#Q.4. How many cars have a mileage greater than 200,000?
len(df[df['Mileage'] > 200000])
```

```python
#Q.5.What is the distribution of cars by transmission type?
df['Transmission'].value_counts()
```

```python
#Q.6 Which car model has the highest price?
df[df['Price'] == df['Price'].max()]
```

```python
#Q.7 What is the average price of cars by fuel type?
df.groupby('Fuel_Type')['Price'].mean()
```

```python
#Q.8 How many cars are from the year 2020 or later?
df[df['Year'] > 2020]
```

```python
#Q.9 What is the average engine size for each brand?
df.groupby('Brand')['Engine_Size'].mean()
```

```python
#Q.10  Which car has the lowest mileage?
df[df['Mileage'] == df['Mileage'].min()]
```

```python
#Q.11 What is the correlation between mileage and price?
df['Mileage'].corr(df['Price'])
```

```python
#Q.12  How many cars have more than 2 owners?
len(df[df['Owner_Count'] > 2])
```

```python
#Q.13 What is the average price of cars by year?
df.groupby('Year')['Price'].mean()
```

```python
#Q.14  Which car has the highest number of doors?
df[df['Doors'] == df['Doors'].max()]
```

```python
#Q.15 What is the average mileage for each fuel type?
df.groupby('Fuel_Type')['Mileage'].mean()
```

```python
#Q.16 How many cars are from the brand 'Toyota'?
len(df[df['Brand'] == 'Toyota'])
```

```python
#Q.17 What is the average price of cars with automatic transmission?
df[df['Transmission'] == 'Automatic']['Price'].mean()
```

```python
#Q.18  Which car has the highest owner count?
df[df['Owner_Count'] == df['Owner_Count'].max()]
```

```python
#Q.19 What is the average engine size for cars with diesel fuel type?
df[df['Fuel_Type'] == 'Diesel']['Engine_Size'].mean()
```

```python
#Q.20 How many cars have a price greater than $10,000?
len(df[df['Price'] > 10000])
```
